```python
%config IPCompleter.greedy=True
%matplotlib inline

# Import the dependencies.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Path
from pathlib import Path
```

```python
# Currently I want to merge two different csv files
# into a larger file to do some analysis on NBA data
#Sources:
#  - https://www.nba.com/stats/players/traditional/?sort=PTS&dir=-1&Season=2019-20&SeasonType=Regular%20Season
#  - https://www.basketball-reference.com/leagues/NBA_2020_per_game.html

stats_nba_file_path = Path("../data/interim/nba-stats-general-traditional-2019-20.csv")
reference_file_path = Path("../data/interim/bbref-general-traditional-2019-20.csv")

stats_nba_df = pd.read_csv(stats_nba_file_path)
reference_df = pd.read_csv(reference_file_path)
```

```python
stats_nba_df = stats_nba_df.sort_values(by="player_name", key=lambda col: col.str.lower())
stats_nba_df
```

```python
stats_nba_df.dtypes
# I want to conver the column of player_id to string.
stats_nba_df["nba_stats_player_id"] = stats_nba_df["nba_stats_player_id"].astype(str)
stats_nba_df.dtypes
```

```python
player_list_name_list = stats_nba_df.loc[:,"player_name"].to_list()
print(len(player_list_name_list))
```

```python
reference_df = reference_df.sort_values(by="player", key=lambda col: col.str.lower())
reference_df
```

```python
player_list = reference_df.loc[:,"player"].to_list()
print(len(player_list))
```

```python
# Checking player names from two different sources to match.

def non_match_elements(list_a, list_b):
    non_match = []
    for i in list_a:
        if i not in list_b:
            non_match.append(i)
    return non_match

non_match_head = non_match_elements(player_list_name_list, player_list)
print(f"No match elements: {non_match_head}")
# print(player_list_name_list)
non_match_tail = non_match_elements(player_list, player_list_name_list)
print(f"No match elements: {non_match_tail}")
# print(player_list)
```

```python
# How to select rows with NaN in particular column?
# df.loc[df['Col2'].isnull()]
na_reference_df = reference_df.loc[reference_df["player"].isnull()]
na_reference_df
```

```python
# Dropped rows where the column Player is equal to NaN
filter_reference_df = reference_df.loc[~reference_df["player"].isin([np.nan])]
filter_reference_df
```

```python
# basketball Reference Data also has duplicate Player values since they were
# on different teams through the season but same person.
filter_reference_df.duplicated(subset=['bbref_player_id'])
```

```python
# basketball Reference Data also has duplicate Player values since they were
# on different teams through the season but same person.

# We need to select the data we will be using for deduplication from our baseketball reference data.

# take the player name, id, team and fields we will use for deduplication from bbref data
# bbref_base_data = filter_reference_df[["Player", "id", "Team", "Field Goal Attempts", "Total Rebounds", "Assists"]].groupby(
#     by="id").apply(deduplicate_traded_players)

# We also need to deduplicate players who played for multiple teams in a single year on

# Basketball reference contains multiple rows for players who have played on multiple teams.
# we only care about the season total for the player so we must deduplicate the rows (selected Team = TOT)
def deduplicate_traded_players(group):
    if len(group) > 1:
        return group[group["team_abbreviation"] == "TOT"]
    return group

deduplicate_df = filter_reference_df[["bbref_player_id", "player", "team_abbreviation", "fga", "reb", "ast"]].groupby(
    by="bbref_player_id").apply(deduplicate_traded_players)
deduplicate_df
```

```python
# take the player name, id, and fields we will use for deduplication from stats.nba.com data
nba_base_data = stats_nba_df[["nba_stats_player_id", "player_name", "team_abbreviation", "fga", "reb", "ast"]]
nba_base_data
```

```python
# Perform a full outer join on the two dataframes. This allows us to get all of the exact matches
name_matches = deduplicate_df.merge(nba_base_data,
                                     left_on=["player", "team_abbreviation", "fga", "reb", "ast"],
                                     right_on=["player_name", "team_abbreviation", "fga", "reb", "ast"], how="outer")

# take all of the exact matches and rename the columns, we only care about player name and id from each source
name_matches_ids = name_matches.dropna()
name_matches_ids = name_matches_ids[["bbref_player_id", "player", "nba_stats_player_id", "player_name"]]
name_matches_ids.columns = ["bbref_name", "bbref_id", "nba_name", "nba_id"]
name_matches
```

```python
# Take all of the rows from the full outer join that have null values. These are the cases where no match was found.
non_matches = name_matches[name_matches.isnull().any(axis=1)]

# take all of the bbref data from non_matches
bbref_non_matches = non_matches[["player", "bbref_player_id", "team_abbreviation", "fga", "reb", "ast"]].dropna()

# take all of the stats.nba data from the non_matches
nba_non_matches = non_matches[["player_name", "nba_stats_player_id", "team_abbreviation", "fga", "reb", "ast"]].dropna()
```

```python
from fuzzywuzzy import fuzz

def check_names_fuzzy_match(row):
    row["name_match"] = fuzz.partial_ratio(row["player"], row["player_name"]) > 60
    return row

possible_matches = bbref_non_matches.merge(nba_non_matches,
                                     left_on=["team_abbreviation", "fga", "reb", "ast"],
                                     right_on=["team_abbreviation", "fga", "reb", "ast"], how="outer").apply(check_names_fuzzy_match, axis=1)

#print out any misses
print(possible_matches[possible_matches["name_match"] != True].head(10))

fuzzy_matches = possible_matches[possible_matches["name_match"]][["bbref_player_id", "player", "nba_stats_player_id", "player_name"]]

# Give the dataframe column names
fuzzy_matches.columns = ["bbref_id", "bbref_name", "nba_id", "nba_name"]
```

```python
# merge our fuzzy matched ids with our exact match ids
all_matches = fuzzy_matches.append(name_matches_ids)

# write to csv
all_matches.to_csv("../data/processed/nba-stats-bbref-general-traditional-2019-20.csv", index=False)
all_matches
```

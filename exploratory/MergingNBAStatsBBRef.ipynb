```python
%config IPCompleter.greedy=True
%matplotlib inline

# Import the dependencies.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Path
from pathlib import Path
```

```python
# Currently I want to merge two different csv files
# into a larger file to do some analysis on NBA data
#Sources:
#  - https://www.nba.com/stats/players/traditional/?sort=PTS&dir=-1&Season=2019-20&SeasonType=Regular%20Season
#  - https://www.basketball-reference.com/leagues/NBA_2020_per_game.html

stats_nba_file_path = Path("../data/interim/player-general-traditional.csv")
reference_file_path = Path("../data/interim/general_2019_20.csv")

stats_nba_df = pd.read_csv(stats_nba_file_path)
reference_df = pd.read_csv(reference_file_path)
```

```python
stats_nba_df = stats_nba_df.sort_values(by="player_name", key=lambda col: col.str.lower())
stats_nba_df
```

```python
player_list_name_list = stats_nba_df.loc[:,"player_name"].to_list()
print(player_list_name_list)
```

```python
reference_df = reference_df.sort_values(by="Player", key=lambda col: col.str.lower())
reference_df
```

```python
player_list = reference_df.loc[:,"Player"].to_list()
print(player_list)
```

```python
# Checking player names from two different sources to match.

def non_match_elements(list_a, list_b):
    non_match = []
    for i in list_a:
        if i not in list_b:
            non_match.append(i)
    return non_match

non_match_head = non_match_elements(player_list_name_list, player_list)
print(f"No match elements: {non_match_head}")
# print(player_list_name_list)
non_match_tail = non_match_elements(player_list, player_list_name_list)
print(f"No match elements: {non_match_tail}")
# print(player_list)
```

```python
# How to select rows with NaN in particular column?
# df.loc[df['Col2'].isnull()] 
na_reference_df = reference_df.loc[reference_df["Player"].isnull()]
na_reference_df
```

```python
# Dropped rows where the column Player is equal to NaN
filter_reference_df = reference_df.loc[~reference_df["Player"].isin([np.nan])]
filter_reference_df
```
